```python
import requests
import numpy as np
import pandas as pd

import csv
```

```python
file_path = '../data/external/worldcities.csv'  
coordinates = []

with open(file_path, newline='', encoding='utf-8') as csvfile:
    csv_reader = csv.DictReader(csvfile)
    for row in csv_reader:
        lat = float(row['lat'])
        lng = float(row['lng'])
        coordinates.append((lat, lng))

# Hiển thị các tuple (lat, lng) trong list coordinates

    
```

```python
# Tạo list để chứa thông tin thời tiết của từng vị trí
weather_info = []

# API key của OpenWeatherMap
api_key = "8402b2eb587d5fb3d3762b8d9e3ea7c9"
i = 0
counter_clear = 0
# Lặp qua từng vị trí trong coordinates và truy vấn thông tin thời tiết
for coord in coordinates:
    url = f"https://api.openweathermap.org/data/2.5/weather?lat={coord[0]}&lon={coord[1]}&appid={api_key}"
    response = requests.get(url)
    print(i , counter_clear)
    i += 1
    
    if response.status_code == 200:
        weather_data = response.json()
        if counter_clear == 1000:
            break
        # Trích xuất thông tin cần thiết từ dữ liệu thời tiết
        country = weather_data.get('sys', {}).get('country', '')
        name = weather_data.get('name', '')
        main_weather = weather_data.get('weather', [{}])[0].get('main', '')
        temp = weather_data.get('main', {}).get('temp', '')
        humidity = weather_data.get('main', {}).get('humidity', '')
        rain = weather_data.get('rain', {}).get('1h', '')
        visibility = weather_data.get('visibility', '')
        wind_speed = weather_data.get('wind', {}).get('speed', '')
        clouds = weather_data.get('clouds', {}).get('all', '')

        if main_weather == 'Clear':
            counter_clear +=1
        # Tạo dictionary chứa thông tin thời tiết của mỗi vị trí
        location_weather = {
            'Country': country,
            'Name': name,
            'Weather': main_weather,
            'Temp': temp,
            'Humidity': humidity,
            'Rain': rain,
            'Visibility': visibility,
            'Wind speed': wind_speed,
            'Clouds': clouds
        }
        
        weather_info.append(location_weather)
    else:
        continue

# Tạo DataFrame từ danh sách thông tin thời tiết của các vị trí
weather_df = pd.DataFrame(weather_info)
print(weather_df)
```
